```
import pandas as pd
import statsmodels.api as sm
url = 'https://raw.githubusercontent.com/AlejandroPSJ2/DataMining_c1/main/Clase1_example1.csv'
df = pd.read_csv(url)
```

```
df.head()
```

```
dfx = df.copy()
dfx = dfx.drop(columns='Diabetes_future',axis=1)
dfx.head()

Y = df['Diabetes_future']
X = dfx
X = sm.add_constant(X)
```

```
mod = sm.Logit(Y,X)
res = mod.fit()
```

```
print(res.summary())
```
